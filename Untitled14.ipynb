```python
colors = ['blue','orange','red']
species = ['Iris-setosa','Iris-versicolor','Iris-virginica']
```

```python
for i in range(3):
    #filter data on each class
    x = df[df['Species'] == species[i]]
    #plot the scatter plot
    plt.scatter(x['SepalLengthCm'],x['SepalWidthCm'],c=colors[i],label=species[i])
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.legend()
```

```python
for i in range(3):
    #filter data on each class
    x = df[df['Species'] == species[i]]
    #plot the scatter plot
    plt.scatter(x['PetalLengthCm'],x['PetalWidthCm'],c=colors[i],label=species[i])
plt.xlabel("PetalLengthCm")
plt.ylabel("PetalWidthCm")
plt.legend()
```

```python
for i in range(3):
    #filter data on each classs
    x=df[df['Species'] == species[i]]
    #plot the scatter plot
    plt.scatter(x['SepalLengthCm'],x['PetalLengthCm'], c = colors[i],label=species[i])
plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
plt.legend()
```

```python
columns_for_correlation = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm','PetalWidthCm']  # Replace with your column names
correlation_matrix = df[columns_for_correlation].corr()
print(correlation_matrix)
```

```python
fig, ax = plt.subplots(figsize=(5, 4))  # Corrected figsize syntax
sns.heatmap(correlation_matrix, annot=True, ax=ax, cmap='coolwarm')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import pickle

#input data
X = df.drop(columns=['Species'])
#output data
Y=df['Species']
#split the data for train and test
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
print("Logistic Regression Accuracy: ",model.score(x_test,y_test)*100)
```

```python
model.fit(x_train.values,y_train.values)
```

```python
print("Accuracy: ",model.score(x_test,y_test)*100)
```

```python
model = KNeighborsClassifier()
model.fit(x_train.values,y_train.values)
print("K-nearest neighbors Accuracy: ",model.score(x_test,y_test)*100)
```
